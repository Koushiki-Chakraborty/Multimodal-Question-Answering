```python
import ipywidgets as widgets
from IPython.display import display, clear_output
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from paddleocr import PPStructure

# 1. Initialize the VL Structure Engine
# We set image_orientation=False to avoid needing extra downloads for now
engine = PPStructure(show_log=True, use_gpu=False, lang='en', layout=True)

# 2. Create the GUI
uploader = widgets.FileUpload(accept='image/*', multiple=False)
output_area = widgets.Output()

def on_upload_change(change):
    with output_area:
        clear_output()
        if not uploader.value:
            return
            
        uploaded_file = uploader.value[0]
        filename = uploaded_file['name']
        with open(filename, "wb") as f:
            f.write(uploaded_file.content)
        
        print(f"🔍 Analyzing: {filename}...")
        
        # 3. Process the Image
        img = cv2.imread(filename)
        result = engine(img)

        # 4. BIG Visualization (30x40 for maximum clarity)
        plt.figure(figsize=(30, 40))
        img_plot = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        
        for line in result:
            # Draw the box around the detected region
            box = np.array(line['bbox'], dtype=np.int32)
            cv2.rectangle(img_plot, (box[0], box[1]), (box[2], box[3]), (0, 255, 0), 5)
            
            # BIG labels for Table vs Text
            label = f"--- {line['type'].upper()} ---"
            plt.text(box[0], box[1]-20, label, color='white', fontsize=28, 
                     fontweight='bold', bbox=dict(facecolor='red', alpha=0.9))

        plt.imshow(img_plot)
        plt.axis('off')
        plt.show()

uploader.observe(on_upload_change, names='value')
display(widgets.VBox([uploader, output_area]))
```

```python
import ipywidgets as widgets
from IPython.display import display, clear_output
from paddleocr import PPStructure
import cv2

# Initialize engine
engine = PPStructure(show_log=False, use_gpu=False, lang='en', layout=True, ocr=True)

uploader = widgets.FileUpload(accept='image/*', multiple=False)
output_area = widgets.Output()

def debug_vl_output(change):
    with output_area:
        clear_output()
        if not uploader.value: return
        
        uploaded_file = uploader.value[0]
        filename = uploaded_file['name']
        with open(filename, "wb") as f:
            f.write(uploaded_file.content)
            
        img = cv2.imread(filename)
        result = engine(img)
        
        print(f"--- RAW DATA STRUCTURE FOR: {filename} ---")
        for i, region in enumerate(result):
            print(f"\nREGION {i}: Type: {region['type']}")
            # Showing the keys helps us see why the 'int()' error happened
            print(f"Keys available in this region: {list(region.keys())}")
            
            if region['type'] == 'table':
                # Table 'res' is usually a dict with 'html' and 'cell_probs'
                print(f"Table data keys: {list(region['res'].keys())}")
                # Print a small sample of the first cell found
                if 'cell_probs' in region['res'] and len(region['res']['cell_probs']) > 0:
                    print(f"Sample Cell Data: {region['res']['cell_probs'][0]}")
            else:
                # Standard text 'res' is a list
                if region['res']:
                    print(f"Sample Text Line: {region['res'][0]}")

uploader.observe(debug_vl_output, names='value')
display(widgets.VBox([uploader, output_area]))
```

```python
import ipywidgets as widgets
from IPython.display import display, clear_output
import cv2
import pandas as pd
from PIL import Image
from paddleocr import PPStructure

# 1. Initialize the PaddleOCR-VL Engine
# We enable 'layout' and 'ocr' to get the structured HTML output
engine = PPStructure(show_log=False, use_gpu=False, lang='en', layout=True, ocr=True)

uploader = widgets.FileUpload(accept='image/*', multiple=False)
output_area = widgets.Output()

def show_original_and_table(change):
    with output_area:
        clear_output()
        if not uploader.value: return
        
        uploaded_file = uploader.value[0]
        filename = uploaded_file['name']
        with open(filename, "wb") as f:
            f.write(uploaded_file.content)
            
        # --- STEP 1: Show the Original Image ---
        print(f"🖼️ Original Image: {filename}")
        display(Image.open(filename))
        print("-" * 50)
        
        # --- STEP 2: Run PaddleOCR-VL ---
        print(f"🔍 Extracting data using PaddleOCR-VL...")
        img = cv2.imread(filename)
        result = engine(img)
        
        # --- STEP 3: Display Extracted Tables ---
        table_found = False
        for region in result:
            if region['type'] == 'table':
                table_found = True
                html_str = region['res']['html'] # The VL-reconstructed structure
                
                try:
                    # Convert HTML structure to a Pandas DataFrame
                    dfs = pd.read_html(html_str)
                    if dfs:
                        print(f"\n📊 Extracted Table Format:")
                        display(dfs[0]) # Show the spreadsheet view
                except Exception as e:
                    print(f"❌ Could not format table: {e}")

        if not table_found:
            print("⚠️ No table structures were detected in this image.")

# Link and display
uploader.observe(show_original_and_table, names='value')
display(widgets.VBox([uploader, output_area]))
```
